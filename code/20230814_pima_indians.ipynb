### 중요한 데이터 추출하기

```python
# plasma를 기준으로 각각 정상과 당뇨가 어느 정도 비율로 분포
plt.hist(x = [df.plasma[df.diabetes == 0], df.plasma[df.diabetes == 1]], bins = 30, histtype = 'barstacked', label = ['noraml', 'diabetes'])
plt.legend()
plt.show()
```

```python
# BMI를 기준으로 각각 정상과 당뇨가 어느 정도 비율로 분포하는지 살펴보자
plt.hist(x = [df.bmi[df.diabetes == 0], df.bmi[df.diabetes == 1]], bins = 30, histtype = 'barstacked', label = ['noraml', 'diabetes'])
plt.legend()
plt.show()
```

### 피마 인디언 당뇨병 예측 실행

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X = df.iloc[:,0:8]   # input_dim : 8
# 당뇨병 여부 Y
y = df.iloc[:,8]
```

```python
# 모델을 설정합니다
model = Sequential()
model.add(Dense(12, input_dim = 8, activation = 'relu', name = 'Dense_1'))  # 입력값 : input_dim
model.add(Dense(8, activation = 'relu', name = 'Dense_2'))
model.add(Dense(1, activation = 'sigmoid', name = 'Dense_3'))    # Dense(1, ...) : 출력값
model.summary()
# param# 108 = (8*12)+12
```

```python
# 모델 컴파일
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
# 모델 실행
history = model.fit(X, y, epochs = 100, batch_size = 5)
```
