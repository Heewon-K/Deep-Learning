## 비전 트랜스포머(Vision Transformer, ViT)



- attention 기법을 사용할 경우 대부분 CNN과 함께 사용되서나 전체 CNN 구조를 유지하면서 CNN의 특정구성요소를 대체하는데 사용
- CNN에 의존하지 않고 이미지 패치의 시퀀스를 입력값으로 사용하는 transformer를 적용

#### 장점
    - transformer구조를 거의 그대로 사용하기 때문에 확장성이 좋다
    - large 스케일 학습에서 매우 우수
    - transfer learning시 CNN보다 훈련에 더 적은 계산 리소스를 사용
#### 단점
    - inductive bias의 부족으로 인해 CNN보다 데이터 많이 요구
    - e.g. 중간 사이즈인 이미지넷으로 강한 정규화 없이 하습에 사용할 경우 유사한 크기의 ResNet보다 성능이 낮음
