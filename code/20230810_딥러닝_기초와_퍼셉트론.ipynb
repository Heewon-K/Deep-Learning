### 2. 퍼셉트론
퍼셉트론(perceptron) 알고리즘

프랑크 로젠브라크(Frank Rosenblatt)가 1957년 고안한 알고리즘

신경망(딥러닝)의 기원이 되는 알고리즘

### 2.1 퍼셉트론이란?
퍼셉트론(인공뉴런): 다수의 신호를 입력으로 받아 하나의 신호를 출력

신호: 흐른다(1), 흐르지 않는다(0) 두 가지 값을 가짐.

입력신호: x1, x2

출력신호: y

가중치: w1, w2(w는 weight의 머리글자)

퍼셉트론에서는 가중치가 클수록 강한 신호를 흘림.

전류에서 말하는 저항. 저항이 낮을 수록 큰 전류가 흐름.

뉴런, 노드: x1, x2, y

입력신호가 뉴런에 보내질 때 가각 고유한 가중치가 곱해짐. (w1x1, w2x2)

뉴런이 활성화: 뉴런에서 보내온 신호의 총합이 정해진 한계를 넘어설 때만 1을 출력

임계값: 정해진 한계. 세타 기호(θ)로 표현


### 2.2 단순한 논리 회로
#### 2.2.1 AND 게이트

```python
def AND(x1,x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = w1*x1 + w2*x2
    if tmp <= theta: # θ보다 작다면 return 0
        return 0
    elif tmp >theta: # θ보다 크다면 return 1
        return 1
```

```python
print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
```

#### 2.2.2 NAND 게이트와 OR 게이트
NAND = not AND. AND 게이트의 출력을 뒤집은 것.

```python
def NAND(x1, x2):
    w1, w2, theta = -0.5, -0.5, -0.7 # 초기값인 가중치 1 가중치2 임계값 설정
    tmp = w1*x1 + w2*x2
    if tmp <= theta:   # θ보다 작다면 0
        return  0
    elif tmp > theta:  # θ보다 크면 1
        return 1
```

```python
print(NAND(0,0))
print(NAND(1,0))
print(NAND(0,1))
print(NAND(1,1))
```

```python
def OR(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.3
    tmp = w1*x1 + w2*x2
    if tmp <= theta:   # θ보다 작다면 0
        return 0
    elif tmp > theta: # θ보다 크다면 1
        return 1
```

```python
print(OR(0,0))
print(OR(1,0))
print(OR(0,1))
print(OR(1,1))
```

#### 2.3.2 가중치과 편향 도입
theta 값을 b라는 편향(bias)로 바꾸는 다른 방식으로 구현

```python
import numpy as np
x = np.array([0,1])      # 입력
w = np.array([0.5,0.5])  # 가중치
b = -0.7                 # 편향
w*x
```

```python
np.sum(w*x)
```

```python
np.sum(w*x) + b
```

```python
def AND(x1, x2):
    x = np.array([x1,x2])      # 입력
    w = np.array([0.5,0.5])  # 가중치
    b = -0.7                 # 편향
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    elif tmp > 0:
        return 1
    
def NAND(x1, x2):
    x = np.array([x1,x2])      # 입력
    w = np.array([-0.5,-0.5])  # 가중치
    b = 0.7                 # 편향
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    elif tmp > 0:
        return 1
    
def OR(x1, x2):
    x = np.array([x1,x2])      # 입력
    w = np.array([0.5,0.5])  # 가중치
    b = -0.2                 # 편향
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    elif tmp > 0:
        return 1
```

```python
print('AND 결과')
print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
print('NAND 결과')
print(NAND(0,0))
print(NAND(1,0))
print(NAND(0,1))
print(NAND(1,1))
print('OR 결과')
print(OR(0,0))
print(OR(1,0))
print(OR(0,1))
print(OR(1,1))
```

### 2.4 퍼셉트론의 한계
#### 2.5.2 XOR 게이트 구현하기

```python
def XOR(x1,x2):
    s1 = NAND(x1,x2)  # 0층
    s2 = OR(x1, x2)
    y = AND(s1, s2)  # 1층
    return y
```

```python
print(XOR(0,0))
print(XOR(1,0))
print(XOR(0,1))
print(XOR(1,1))
```
