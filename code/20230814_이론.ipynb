## 퍼셉트론과 인공지능의 시작
### 1. 인공지능의 시작으로 알린 퍼셉트론
    - 인간의 뇌 : 뉴론과 시냅스 --> 시냅스에서 화학물질이 나와 전위 변화를 일으키는데, 전위가 임계값을 넘지 못하면 아무것도 하지 않고 임계값을 넘을때만 다름 뉴런으로 신호가 전달된다.
    - 이 매커니즘은 로지스틱 회귀와 닮아있다.
    - 입력값이 활성화 함수에 의해 이전한 수준을 넘으면 참, 그렇지 않으면 거짓.
    - ==> 이렇게 시작된 연구가 인공 신경망(artificial neural network)
    - 57' 프랑크 로젠블랏이 이 개념을 실제 장치로 만들어 보인 것이 퍼셉트론(perceptron)
#### * 퍼셉트론
- 입력갑을 여러개 받아 출력을 만드는데, 입력값에 가중치를 조절할 수 있게 만들어 최초로 '학습'을 하게 했다.
- **아달라인(adaline)**: 여기에 경사 하강법을 도입해 최적의 경계선으 그릴 수 있게 한것이 아달라인
    - 이후 소프트 벡터 머신(Support Vector Machine)등 머신 러닝의 중요한 알고리즘으로 발전. 
    - 이 중 시그모이드를 활성화 함수로 사용한 것이 로지스틱 회귀
    

### 2. 퍼셉트론의 한계
- 하나의 직선으로는 다음과 같이 위치하는 흰색 점과 검은색 점을 구분할 수 없다.


### 3. XOR 문제
- AND게이트는 x1과 x2 둘 다 1일때 결과값이 1
- OR게이트는 둘 중 하나라도 1이면 결과값이 1
- XOR게이트는 둘 중 하나만 1일때 1
- XOR게이트는 직선을 그어 구분 할 수 없다
- ==> 퍼셉트론으로 인공지능이 가능하리라 꿈꾸었지만 간단한 XOR문제 조차 해결할 수 없었다.
- Marvin Minsky의 69년도 논문 이후 인공지능 연구가 침체기를 겪게 됨
#### 해결
- 다중 퍼셉트론(multilayer perceptron)과 오차역전파(back propagation)의 개발로 해결되었다

## 다중 퍼셉트론의 등장
: 차원을 달리하면 해결되지 않을까?



- 종이를 휘어 주어 선 두개를 동시에 긋는 방법
- 이것을 XOR문제에 적용하면 '퍼셉트론 두 개를 한번에 계산'
- ==> 퍼셉트론 두 개를 각각 처리하는 은닉층(hidden layer)를 만듦



- 다중 퍼셉트론의 내부



- x1과 x2는 입력값, 각 값에 가중치(w)를 곱하고 바이어스(b)를 더해 은닉층으로 이동 
- 은닉층의 중간 정거장을 노드(node)라고 하며 n1, n2로 표시
- 은닉층에 들어온 값들은 활성화 함수를 통해 다음으로 보내진다
- 출력층에서 활성화 함수를 사용해 y 예측값을 출력

### XOR문제 해결

- 숨어있는 노드 두개를 둔 다중 퍼셉트론을 통해 XOR문제 해결

## 오차 역전파
: XOR문제를 다중 퍼셉트론으로 해결했지만 다른 문제를 만났다
- *은닉층에 포함된 가중치를 업데이트할 방법이 없다!*
- 긴 인공지능의 겨울을 지나 **오차 역전파**라는 방법을 만나 해결

### 딥러닝의 태동, 오차 역전파
- 앞서 XOR 문제를 해결했지만 입력값과 출력값을 알고 있는 상태에서 가중치와 바이어스를 넣었으므로 '모델링'이라고 할 수 없다.
- 우리가 원하는 것은 데이터를 통해 스스로 가중치를 조절하는 '학습'의 실현!
- **경사 하강법**은 임의의 가중치를 선언하고 결과값을 이용해 오차를 구한 후 이 오차가 최소인 지점(미분했을때 기울기가 0인 지점)으로 계속하여 조금씩 이동하는 것.
- 은닉층이 생기면서 두번의 경사 하강법을 실행해야한다



- 먼저 한번의 순전파(1)가 일어남
- 이를 통해 각 가중치의 초깃값이 정해짐
- 초기 가중치로 만들어진 값과 실제값을 비교해 출력층의 오차를 계산
- 계산된 출력층의 오차를 최소화시키기 위해 첫번째 가중치를 수정하는 과정(2)과 두번째 가중치를 수정하는 과정(3)이 이어짐
- 예) w31을 업데이트 하려고 한다면 오차 공식을 구하고 w31에 대해 편미분을 해야한다. 합성함수의 미분이므로 체인 룰을 적용해 편미분을 구하고 이를 통해 w31을 업데이트할 수 있다.
- 예2) w11을 업데이트 하려고 할때 마찬가지로 오차 공식을 구하고 편미분으 하면 된다. **하지만** 출력층의 결과와 실제 값을 비교해 오차를 구해야하는데 은닉층은 겉으로 들어나지 않으므로 그 값을 알 수 없음...
- ==> 이 문제는 다시 출력틍의 오차를 이용하는 것으로 해결
- w31의 경우 yo1의 오차만 필요했지만, w11은 yo1와 yo2가 모두 관여되어 있다.



- 위 공식에서 밑줄 친 부분은 'out(1-out)'의 형태 : 델타식이라고 한다
- 은닉층의 숫자가 늘어도 이러한 형태가 계속해서 나타나게 되므로, 이를 이용해 깊은 층의 계산도 할 수 있게 됨.
- ==> 이렇게 깊은 층을 통해 학습할 수 있는 계기가 마련되면서 **딥러닝**이 태동하게 되었다
### 활성화 함수와 고급 경사 하강법
: 델타식을 통해 깊은 신경망의 계산이 가능해졌지만 아직 한가지 문제가 더 남아 있다.
==> * 기울기 소실의 문제!



- 이는 활성화 함수로 사용된 시그모이드 함수의 특성 때문!
- 여러층을 거칠수록 기울기가 사라져 가중치를 수정하기 어려워진다



- ==> 이를 해결하기 위해 제프리 힌튼교수가 렐루(ReLU)함수를 제시!



- 활성화 함수로 렐루를 쓰면 여러번 오차 역전파가 진행되어도 맨 처음 층까지 값이 나ㅏㅁ아 있게된다
- 학습 : 오차를 최소화하는 가중치를 찾는 과정
- 출력층에서 알아낸 오차가 역전파를 통해 입력층까지 거슬로 올라가면서 잘못된 가중치들을 수정할 수 있게 되어 더 깊은 층을 쌓아 올리는 것이 가능해졌다
- 렐루 이후에도 하이퍼볼릭 탄젠트(hyperbolic tangent)함수나 소프트플러스(softplus)함수등 좀 더 나은 활성화 함수를 만든기 위한 노력이 이어지고 있음

### 속도와 정확도문제 해결을 위한 고급 경사 하강법
: 경사 하강법은 정확하게 가중치를 찾아가지만, 계산량이 매우 많다는 단점이 있다!
==> 이러한 점을 보완한 고급 경사 하강법이 등장, 딥러닝의 발전 속도가 더 빨라졌다.

#### 1) 확률적 경사 하강법(Stochastic Gradient Descent, SGD)
- 경사 하강법은 한번 업데이트를 할 때마다 전체 데이터를 미분하므로 속도가 느릴 뿐만 아니라 최적의 해를 찾기 전에 최적화 과정이 멈출 수 있다.
- SGD는 전체 데이처를 사용하는 것이 아니라, 랜덤하게 추출한 일부 데이터만 사용하기 떄문에 빠르고 더 자주 업데이트할 수 있다는 장점이 있다
- 랜덤한 일부 데이터를 사용하는 만큼 확률적 경사 하강법은 중간 결과의 진폭이 크고 불안정해 보일 수 있음
- 속도가 확연히 빠르면서도 최적 해에 근사한 값을 찾아낸다는 장점 덕분에 경사 하강법의 대안으로 사용되고 있다.



#### 2) 모멘텀(momentum) : 관성, 탄력, 가속도
- 모멘텀 확률적 경사 하강법(모멘텀 SGD) : 경사 하강법에 탄력을 더해 주는 것
- 경사 하강법과 마찬가지로 매번 기울기를 구하지만 이를 통해 오차를 수정하기 전 바로 앞 수정 값과 방향(+,-)을 참고해 같은 방향으로 일정한 비율만 수정되게 하는 방법
- 수정 방향이 양수(+)방향으로 한 번, 음수(-)방향으로 한 번 지그재그로 일어나는 형상이 줄어들고, 이전 이동값을 고려헤 일정 비율만큼 다음 값을 결정하므로 관성 효과를 낼 수 있다.



- ==> 지금은 정확도와 속도를 모두 향상시킨 아담(adam)이라는 고급 경사 하강법이 가장 많이 쓰인다



#### 옵티마이저
: 이렇게 오차를 최소화 하는 경사 하강법들을 '옵티마이저'라고 부른다.
- 텐서플로에 포함되어 있는 opimizers라는 객체에 이름을 적어 주는 것만으로 손쉽게 실행
- 앞에서 소개된 시그모이드, 렐루 등 활성화 함수도 activation이라는 객체에 이름을 적어 손쉽게 실행
